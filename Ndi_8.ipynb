```python
import pandas as pd
import seaborn as sns
```

```python
avo = pd.read_csv('avocado.csv')
avo.head(5)
```

```python
avo.info()
```

```python
avo['Date']=pd.to_datetime(avo['Date'])
```

```python
avo.info()
```

```python
avo2 = avo.query('year == 2015 and type == "conventional" and region == "Albany"')[['Date', 'Total Bags', 'Small Bags']]
```

```python
avo2.head()
```

```python
avo_melt = avo.melt(id_vars = 'Date',\
                    value_vars = ['Total Bags', 'Small Bags'],\
                    var_name = 'Bags',\
                    value_name = 'count')
```

```python
avo_melt
```

```python
sns.lineplot(data=avo_melt, x='Date', y='count', 
            hue='Bags')
```

```python
avo['Sales Volume'] = pd.qcut(avo['Total Volume'],\
                             q=4, \
                             labels=('Poor', 'Modest', 'Good', 'Excellent'))
avo.head()
```

```python
#sns.catplot(data=avo, kind='count', x='year', hue='Sales Volume')
sns.countplot(data=avo, x='year', hue='Sales Volume')
```

```python
avo_melt3 = avo.melt(id_vars = 'year',\
                    value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'],\
                    var_name = 'Bags',\
                    value_name = 'Count')
avo_melt3
```

```python
sns.relplot(data=avo_melt3, kind='line', x='year', y='Count', hue='Bags')
```

Base on the plot above, it is clear that the Small Bags size is the best seller.
